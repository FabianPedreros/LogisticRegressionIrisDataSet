Supervised learning, as the name indicates, has the presence of a supervisor as a teacher. Basically supervised learning is when we teach or train the machine using data that are well labeled. This means some data is already tagged with the correct answer. After that, the machine is provided with a new set of examples(data) so that the supervised learning algorithm analyses the training data(set of training examples) and produces a correct outcome from labeled data.

Classification

A classification problem is when the output variable is a category, such as “Red” or “blue”, “disease” or “no disease”



**Supervised classifications algorithms**

**Logistic Regression**

Is a Machine Learning algorithm that is used for classification problems and is based on the concept of probability. It is used when the dependent variable (target) is categorical. It is widely used when the classification problem at hand is binary; true or false, yes or no, etc. Logistics regression uses the sigmoid function to return the probability of a label.

## <a id='9'>9. Data Preparation</a>

In our data we have the variable 'specie' and this is qualitative, we have to make a convertion in order the algorithm read this categorization into numerical values such as ones and zeros, in this case 1, 2 and 3.

We have understood our data and review if this have a normal distribution or not, according to what we have seen, we do not need to perform transformations to the quantitative data, since they do not have behavior that deviates from a normal distribution when reviewed by species.

So the unique transformation we need is to encode the categorical variable 'specie'

```python
# For specie we associate the values
df_dummy = df
df_dummy['specie'] = df_dummy['specie'].replace({'setosa':0, 'versicolor':1, 'virginica':2})
```

```python
# Separate predictors from the target
X = df_dummy.drop(columns = ['specie'])
Y = df_dummy['specie']
```

```python
# Import of the libraries for the division of the dataset into train and test
from sklearn.model_selection import train_test_split
```

```python
# Split data into trainig and test data sets
# We are using a pseudo random number with root = 123, this helps us with replication.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 1234)
```

```python
# Check the shapes of the X_train and X_test
X_train.shape, X_test.shape
```

## <a id='10'>10. Modeling</a>

```python
# Import of the libraries for the division of the dataset into train and test, as well as for the evaluation of algorithms.
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix
```

```python
# Instantiate the model
model = LogisticRegression()
```

```python
# Fit the model
model.fit(X_train, Y_train)
```

```python
kfold = KFold(n_splits = 5, random_state = 42)
cv_res = cross_val_score(model, X_train, Y_train, scoring = 'accuracy', cv = kfold)
print("Accuracy: %0.2f" % (cv_res.mean()*100) + '%')
```

The accuracy of the model with the train data set is 95.24%

```python
# Predict the species for the test data set
Y_pred = model.predict(X_test)
# Summary of the predictions made by the classifier
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
# Accuracy score
print('accuracy is', (accuracy_score(Y_pred, Y_test)*100))
```

The accuracy of the model predicting the species with the test data set is 97.7%

```python
# Confusion matrix generation
cf_matrix = confusion_matrix(Y_test, Y_pred)
plt.figure(figsize=(6,6))
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')
ax.set_title('Confusion Matrix Logistic Regression\n\n');
ax.set_xlabel('\nPredicted Iris Specie')
ax.set_ylabel('Actual Iris Specie ');
ax.xaxis.set_ticklabels(['Setosa', 'Versicolor', 'Virgnica'])
ax.yaxis.set_ticklabels(['Setosa', 'Versicolor', 'Virgnica'])
plt.show();
```

In the confusion matrix we can see how the model tries to predict 45 species, and how it accurately predicts 16 setosa as setosa, 16 versicolor as versicolor, 12 virginica as virginica and fails in one prediction, the model says it is a virginica when in fact it is a versicolor.

We can end our exercise here, because we have found a really good model that classifies iris species accurately, and we don have many regressors to run a backward elimination.
